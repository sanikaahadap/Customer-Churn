```python
def print_unique_col_values(df):
    for column in df:
        if df[column].dtypes=='object':
          print(f'{column}:{df[column].unique()}')
```

```python
print_unique_col_values(df1)
```

```python
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
print_unique_col_values(df1)
```

```python
y_n_cols=['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in y_n_cols:
    df1[col].replace({'Yes':1,'No':0},inplace=True)
```

```python
for col in df1:
    print(f'{col}:{df1[col].unique()}')
```

```python
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df1['gender'].unique()
```

```python
df1
```

```python
df2=pd.get_dummies(data=df1,columns=['InternetService','Contract','PaymentMethod'])
df2
```

```python
df2.columns
```

```python
df2.sample(5)
```

```python
df2.dtypes
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
for col in df2:
    print(f'{col}: {df2[col].unique()}')
```

```python
X=df2.drop('Churn',axis='columns')
Y=df2['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train[:10]
```

```python
len(X_train.columns)
```

```python
import tensorflow as tf
from tensorflow import keras

model=keras.Sequential([
    keras.layers.Dense(20,input_shape=(26,),activation='relu'),
    keras.layers.Dense(1,activation='sigmoid'),
])

model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])

model.fit(X_train,Y_train,epochs=150)
```

```python
model.evaluate(X_test,Y_test)
```

```python
yp=model.predict(X_test)
yp[:5]
```

```python
Y_test[:5]
```

```python
y_predicted=[]
for i in yp:
    if i>0.5:
        y_predicted.append(1)
    else:
        y_predicted.append(0)
```

```python
y_predicted[:15]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(Y_test,y_predicted))
```

```python
import seaborn as sns
con_matrix=tf.math.confusion_matrix(labels=Y_test,predictions=y_predicted)

plt.figure(figsize=(10,7))
sns.heatmap(con_matrix, annot=True,fmt='d')
plt.xlabel('Predictions')
plt.ylabel('Observations')
```
